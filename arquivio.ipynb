```python
import pandas as pd
import time
import urllib.error

def carregar_clientes(sheet_name, tentativas=3, delay=5):
    for i in range(tentativas):
        try:
            df = pd.read_excel(URL_CLIENTES, sheet_name=sheet_name)
            return df
        except urllib.error.HTTPError as e:
            if e.code == 503:
                print(f"Serviço indisponível, tentativa {i+1}/{tentativas}. Aguardando {delay} segundos...")
                time.sleep(delay)
            else:
                raise
    raise Exception("Serviço indisponível após várias tentativas.")
```
